criteria:

| Marks         | Grade | 
|---------------|-------|

2. Write a program to accept the cost price of a bike and display the road tax to be paid according to the
 
 following criteria:
 
| Tax | Cost Price(in Rs)      |
|------|-----------------------|
| 15 % | >100000               |
| 10%  | >50000 and <= 100000  |
| 5%   | <= 50000              |

price = int(input("Enter the cost price of the bike: "))

if price > 100000:

road_tax = 0.15 * price

print("Road tax to be paid: 15% of", price, "=", road_tax)

elif price > 50000 and price <= 100000:

road_tax = 0.10 * price

print("Road tax to be paid: 10% of", price, "=", road_tax)

elif price <= 50000:

road_tax = 0.05 * price

print("Road tax to be paid: 5% of", price, "=", road_tax)

else:
print("Invalid input! Please enter a valid cost price.")

|--------|-----------|

Q.6 Use nested while loops to print 3 different patterns.

n = 5  # Number of rows

# Pattern 1: Triangle Pattern
i = 1
while i <= n:
    j = 1
    while j <= i:
        print("*", end=" ")
        j += 1
    print()
    i += 1
    
n = 4  # Number of rows and columns

# Pattern 2: Square Pattern
i = 1
while i <= n:
    j = 1
    while j <= n:
        print("*", end=" ")
        j += 1
    print()
    i += 1
    
n = 5  # Number of rows

# Pattern 3: Pyramid Pattern
i = 1
while i <= n:
    # Print spaces
    j = 1
    while j <= n - i:
        print(" ", end=" ")
        j += 1

    # Print stars
    j = 1
    while j <= 2 * i - 1:
        print("*", end=" ")
        j += 1

    print()
    i += 1
